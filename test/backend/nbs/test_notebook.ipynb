```python
# No proper headers or organization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# api_key = 'sk_live_secret123'  # DANGER: credentials in notebook
```

```python
# Load data from hardcoded path
data = pd.read_csv('/Users/john/Desktop/data.csv')
print(data.head())
```

```python
# Scale first, then split - DATA LEAKAGE!
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data.iloc[:, :-1])
y = data.iloc[:, -1]

# Now split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
```

```python
# No reproducibility
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Evaluate on training data only
from sklearn.metrics import accuracy_score
train_pred = model.predict(X_train)
acc = accuracy_score(y_train, train_pred)
print(f'Accuracy: {acc}')
```

```python
# Repeated plotting logic
plt.figure(figsize=(10, 6))
plt.plot(y_train)
plt.title('Training Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(y_test)
plt.title('Test Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
